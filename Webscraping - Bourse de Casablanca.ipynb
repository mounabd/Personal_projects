## Extraction des données relatives aux actions

```python
import requests
from bs4 import BeautifulSoup

url_liste = 'http://www.casablanca-bourse.com/bourseweb/Liste-Societe.aspx?IdLink=20&Cat=7'
res = requests.get(url_liste)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
soup.title
```

```python
#<tr class="arial11gris">

tr = soup.find('tr', attrs={'class', 'arial11gris'}) #l'identifiant de la table qu'on recherche soit la liste des états
print(tr)
all_actions = soup.findAll('tr',  attrs={'class', 'arial11gris'} )
print(all_actions)
```

```python
liste_societe = []

for item in all_actions:
    print(item.find_all('td'))
    a = item.find('a')
    if a != None :
        liste_societe.append(a.get_text())
```

```python
print("Nombre de société quotées:",len(liste_societe))
print("Liste de toutes les sociétés quotées en bourse: \n \n",liste_societe)
```

```python
liste_links = []

for item in all_actions:
    link = item.find('a')['href']
    liste_links.append('http://www.casablanca-bourse.com/bourseweb/'+link)
```

```python
print("Nombre de liens :",len(liste_links))
print("Liste des liens par action : \n \n ",liste_links)
```

```python
for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    all_cours = soup.findAll('tr',  attrs={'class', 'tr_blue'})
    
    for item in all_cours:
        print(item.find('span'))
```

### Extraction des cours, des variations journalières, du volume et de la capitalisation 

```python
liste_cours = []

for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    table = soup.findAll('table', attrs = {'class','txt_table'})
    print(stock)
    
    for index,item in enumerate(table) :
        print(item)
        cours = item.find('span' , id = "SocieteCotee1_ctl16_LBCrsCourant" )
        
        if cours != None :
            liste_cours.append(cours.get_text())
            
            
%time
```

```python
liste_capi = []

for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    capi = soup.find('span', id = "SocieteCotee1_ctl16_LBCapitalisation")
    liste_capi.append(capi.get_text())
    print(stock)
    print(capi)
```

```python
liste_variation = []

for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    var = soup.find('span', id = "SocieteCotee1_ctl16_LBVariation")
    liste_variation.append(var.get_text())
    print(stock)
    print(var)
    
%time
```

```python
liste_volume = []

for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    vol = soup.find('span', id = "SocieteCotee1_ctl16_LBVolume")
    liste_volume.append(vol.get_text())
    print(stock)
    print(vol)
    
%time
```

```python
liste_volume_titres = []

for stock, url in zip(liste_societe, liste_links) : 
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'html.parser')
    vol = soup.find('span', id = "SocieteCotee1_ctl16_LBCumulTitreEch")
    liste_volume_titres.append(vol.get_text())
    print(stock)
    print(vol)
    
%time
```

```python
df_total = pd.DataFrame({ 'nom_action' : liste_societe,
                         'cours' : liste_cours,
                         'variation' : liste_variation,
                        'volume': liste_volume, 
                         'volume_titres' : liste_volume_titres,
                          'capitalisation' : liste_capi})
```

```python
df_total[['cours', 'volume', 'volume_titres' , 'capitalisation']]= df_total[['cours', 'volume', 'volume_titres' , 'capitalisation']].apply(lambda x : x.str.replace('Â', ''))
df_total[['cours', 'volume','volume_titres' ,  'capitalisation']]= df_total[['cours', 'volume','volume_titres' ,  'capitalisation']].apply(lambda x : x.str.replace(',','.'))
df_total[['volume','volume_titres' ]]= df_total[['volume', 'volume_titres' ]].apply(lambda x : x.str.replace('-','0'))
df_total[['cours', 'volume','volume_titres' ,  'capitalisation']]= df_total[['cours', 'volume','volume_titres' , 'capitalisation']].apply(lambda x : x.str.replace(u'\xa0', u''))
```

```python
df_total[['cours', 'volume', 'capitalisation','volume_titres' ]]= df_total[['cours', 'volume', 'capitalisation','volume_titres']].astype(float)                                                                           
                                                                                                                                                                                                                                                       
df_total
```
